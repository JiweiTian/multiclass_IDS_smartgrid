```python
import pandas as pd
from scipy.io import arff
import numpy as np
from sklearn.model_selection import train_test_split
import erlc as e
import dataPipeline as dp
from sklearn.metrics import f1_score, accuracy_score, matthews_corrcoef
```

# 1.0 Data

```python
dataPipeline = dp.DataPipeline()
data = dataPipeline.loadData()
X, y = dataPipeline.dataProc(data)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# 2.0 Model
## 2.1 Initializing and Training

```python
model = e.ERLC()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')
mcc = matthews_corrcoef(y_test, y_pred)
print("Accuracy = {}, F1-score = {}, MCC = {}".format(acc, f1, mcc))
```

Below we save the model for later use. The default directory is "saved_model" in the project folder. But this can be changed as an input to save_model called "save_path".

```python
model.save_model(save_path = 'saved_model/')
```

## 2.2 Loading Saved Model

```python
loaded_model = e.ERLC()
loaded_model.load_model(save_path = 'saved_model/')
```

```python
y_pred = loaded_model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')
mcc = matthews_corrcoef(y_test, y_pred)
print("Accuracy = {}, F1-score = {}, MCC = {}".format(acc, f1, mcc))
```
