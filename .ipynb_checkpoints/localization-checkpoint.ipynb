```python
import pandas as pd
from scipy.io import arff
import numpy as np
from scipy import stats
from sklearn.model_selection import train_test_split

from sklearn import preprocessing

import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Data

```python
print('Loading dataset #1')
data = pd.DataFrame(arff.loadarff('input/power_multiclass/data1.arff')[0])

for i in range(2,16):
    print('Loading dataset #{}'.format(i))
    dataTemp = pd.DataFrame(arff.loadarff('input/power_multiclass/data{}.arff'.format(i))[0])
    data = pd.concat([data,dataTemp],axis=0)

print("Finished Loading. Final Size = {}".format(data.shape))
```

```python
def dataProc(df):
    '''
    This function takes a dataframe and splits it into data and labels, proccesses them into numpy arrays and
    splits them into training, validation, and testing data and labels.
    '''
    df = df.astype(np.float64)
    label = df['marker'].astype(int)
    df = df.drop(['marker'], axis=1)
    
    df = df.drop(['snort_log1','snort_log2','snort_log3','snort_log4',
                'control_panel_log1','control_panel_log2','control_panel_log3','control_panel_log4',
                'relay1_log','relay2_log','relay3_log','relay4_log'], axis=1)
    
    df = df.reset_index()
    
    df = df.replace(-np.inf, 0)
    df = df.replace(np.inf, 0)

    
    # Converting to arrays
    X = np.asarray(df)
    y =  np.asarray(label)
    
    # Scaling data
    scalar = preprocessing.MinMaxScaler()
    X = scalar.fit_transform(X)
    
    return X,y
```

```python
X,y = dataProc(data)
print('Size of X: {}'.format(X.shape))
print('Size of y: {}'.format(y.shape))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Localization

```python
def getLocalizationStats (X, y, nomral_label = 41):
    '''
    This function gets the localization statistics, which are the mean and standard deviation of each measurement
    under normal observations. It returns means and stds to be used in percent difference calculation for attack
    localization.
    
    inputs
    --------
    X: the training data
    
    outputs
    --------
    means: The mean of each measurement in the training data
    stds: The standard deviation of each measurement in the training data
    '''
    
    means = []
    stds = []
    
    for i in range(len(X)):
        mean_temp = np.mean(X[i])
        std_temp = np.std(X[i])
        
        # Replacing zeros with small numbers
        if (mean_temp == 0):
            mean_temp = 0.000001
        if (std_temp == 0):
            std_temp = 0.000001   
            
        # Adding them to a list    
        means.append(mean_temp)
        stds.append(std_temp)
    
    means = np.asarray(means)
    stds = np.asarray(stds)
    
    return means, stds


def getPD(sample_np, normal_means, normal_stds):
    '''
    This function takes in a sample as a numpy array and calculates the percent difference from the mean
    of every measurement from the normal data.
    
    inputs
    -------
    sample_np: Sample in numpy format
    normal_means: means of the normal data (from getLocalizationStats)
    normal_stds: standard deviation of the normal data (from getLocalizationStats)
    
    outputs
    -------
    p_diff: percent difference of the sample
    '''
    
    p_diff = []

    for i in range(len(sample_np)):
        comparison_value = abs(normal_means[i]) + abs(normal_stds[i])
        percent_diff = abs((sample_np[i] - comparison_value)/comparison_value)
        p_diff.append(percent_diff)
    
    p_diff = np.asarray(p_diff)
    return p_diff
```

```python
def getAveragePDforEachAttack(X_test, y_test, normal_means, normal_stds):
    
    
    p_diff_avg_final = []
    
    for i in np.unique(y_test):
        print("Iteration i = {}".format(i))
        p_diff_temp = getPD(X_test[y_test==i], normal_means, normal_stds)
        p_diff_temp = np.asarray(p_diff_temp)
        
        print("p_diff_temp shape = {}".format(p_diff_temp.shape))
        
        p_diff_avg = []
        for j in range(p_diff_temp.shape[1]):
            avg_temp = np.mean(p_diff_temp[j])
            p_diff_avg.append(avg_temp)
            
        p_diff_avg = np.asarray(p_diff_avg)
        print("p_diff_avg shape = {}".format(p_diff_avg.shape))
        p_diff_avg_final.append(p_diff_avg)
        
    
    p_diff_avg_final = np.asarray(p_diff_avg_final)
    
    return p_diff_avg_final 
```

First we get localization stats from training data

```python
normal_means, normal_stds = getLocalizationStats(X_train[y_train==41])
normal_means.shape
```

```python
p_diff = getPD(X_test, normal_means, normal_stds)
np.asarray(p_diff).shape
```

```python
p_diff_final = getAveragePDforEachAttack(X_test, y_test, normal_means, normal_stds)
p_diff_final.shape
```
